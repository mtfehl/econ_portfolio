```matlab
%%% Simulation Code %%%
%----------------------
clear

% Task: build 1000 monte carlo simulations

rng(14) % set seed
sims = 1000; % num of sims

% to create conways game of life, our universe is a matrix.
% the non-zero elements are the population

life = zeros(4); % initalize our 4x4 grid of zeros

rows = [1, 1, 2]; % set our agents
cols = [1, 2, 2];
idx = sub2ind(size(life), rows, cols);
life(idx) = 1;

% visualize our non-zero points (agents)
spy(life)
```
